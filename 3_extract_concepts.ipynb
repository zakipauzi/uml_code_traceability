### Test sample UML concept extraction

````python
import torch 
torch.cuda.is_available()
````

````python
from transformers import AutoModel, AutoTokenizer

filepath_model = "C:/Users/biadge/MiniCPM-o-2_6"  # path to the model
model = AutoModel.from_pretrained(filepath_model, 
                                  trust_remote_code=True, 
                                  use_safetensors=True,
                                  attn_implementation='sdpa', 
                                  torch_dtype=torch.bfloat16, 
                                  init_vision=True,
                                  init_audio=True,
                                  init_tts=True)

model = model.eval().cuda()

tokenizer = AutoTokenizer.from_pretrained(filepath_model, trust_remote_code=True)
model.init_tts()
````

````python
from PIL import Image
import ast

with open("concept_extract_prompt.txt", "r") as file:
    prompt = file.read().strip()

def run_extract_concepts_local_llm(model, tokenizer, file) -> dict | str:
    try:
        image = Image.open(file).convert('RGB')

        question = prompt
        msgs = [{'role': 'user', 'content': [image, question]}]
        res = model.chat(
            msgs=msgs,
            tokenizer=tokenizer
        )
    except Exception as e:
        return str(e)

    # remove the ```json from the response
    res = res.replace("```json", "")
    res = res.replace("```", "")

    # remove the first line
    res = res[res.find('\n') + 1:]

    # remove the last line
    res = res[:res.rfind('\n')]

    try:
        # convert string into dictionary
        res = dict(ast.literal_eval(res))
    except Exception as e:
        return str(res)

    return res
````

````python
# save json string into a json file

import json

with open("uml_example.json", "w") as file:
    json.dump(run_extract_concepts_local_llm(model, tokenizer, "uml_example.png"), file)
````

### Outlier set 2 projects

````python
outlier_proj_names = ['apache/wicket/','stritti/log4js/','abulka/pynsource/']

out_sampled_cd_df_stars = pd.read_csv('out_sampled_cd_df_stars.csv')
out_sampled_cd_df_stars['Class_Diagram_Files'] = out_sampled_cd_df_stars['Class_Diagram_Files'].apply(lambda x: ast.literal_eval(x))

out_sampled_cd_df_stars = out_sampled_cd_df_stars[out_sampled_cd_df_stars['Project'].isin(outlier_proj_names)]
out_sampled_cd_df_stars = out_sampled_cd_df_stars[['Project',	'Class_Diagram_Files',	'No_of_Class_Diagrams']]
# explode the class diagram files
out_sampled_cd_df_stars = out_sampled_cd_df_stars.explode('Class_Diagram_Files')
out_sampled_cd_df_stars
````

````python
outlier_proj_names = [outlier_proj_names[-1]]
outlier_proj_names
````

````python
# For every project, we will extract the concepts from the class diagrams. The class diagrams can be found in the 'class_diagrams' folder.
# The extracted concepts will be saved in the 'extracted_concepts' folder.

import os
from tqdm import tqdm
import json

for proj_name in outlier_proj_names:
    proj_df = out_sampled_cd_df_stars[out_sampled_cd_df_stars['Project'] == proj_name]
    class_diagrams = proj_df['Class_Diagram_Files'].tolist()
    for class_diagram in tqdm(class_diagrams, desc=f"Processing Diagrams for {proj_name}, {class_diagram}", leave=False):
        file_path = os.path.join("class_diagrams", proj_name, os.path.basename(class_diagram))
        if not os.path.exists(file_path):
            print(f"File not found: {file_path}")
            continue
        os.makedirs("extracted_concepts", exist_ok=True)
        project_folder = os.path.join("extracted_concepts", proj_name.strip("/"))
        os.makedirs(project_folder, exist_ok=True)
        with open(os.path.join(project_folder, f"{os.path.basename(class_diagram)}.json"), "w") as file:
            output = run_extract_concepts_local_llm(model, tokenizer, file_path)
            try:
                json.dump(output, file)
            except Exception as e:
                print(f"Error json-ifying {file_path}: {e}")
                # save output as bytes to file
                with open(os.path.join(project_folder, f"{os.path.basename(class_diagram)}.txt"), "w") as file:
                    file.write(output)
                continue
````
