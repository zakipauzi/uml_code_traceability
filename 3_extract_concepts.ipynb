```python
import pandas as pd
import ast

out_sampled_cd_df = pd.read_csv('out_sampled_cd_df.csv')
out_sampled_cd_df['Class_Diagram_Files'] = out_sampled_cd_df['Class_Diagram_Files'].apply(lambda x: ast.literal_eval(x))
out_sampled_cd_df['Archived_Class_Diagram_Files'] = out_sampled_cd_df['Archived_Class_Diagram_Files'].apply(lambda x: ast.literal_eval(x))
out_sampled_cd_df
```

```python
# show the projects where there is an increased number of diagrams

out_sampled_cd_df['has_increase'] = out_sampled_cd_df['No_of_Class_Diagrams'] > out_sampled_cd_df['No_of_Archived_Class_Diagrams']
out_sampled_cd_df = out_sampled_cd_df[out_sampled_cd_df['has_increase']].drop(columns="has_increase")
out_sampled_cd_df.to_csv("increased_projects.csv", index=False)
out_sampled_cd_df
```

```python
def remove_dup(curr_list, old_list):
    return set(curr_list) - set(old_list)
    
out_sampled_cd_df['New_Class_Diagram_Files'] = out_sampled_cd_df.apply(lambda row: list(remove_dup(row['Class_Diagram_Files'], row['Archived_Class_Diagram_Files'])), axis=1)
out_sampled_cd_df['No_of_New_Class_Diagram_Files'] = out_sampled_cd_df['New_Class_Diagram_Files'].apply(len)
out_sampled_cd_df.to_csv("increased_projects.csv", index=False)
```
