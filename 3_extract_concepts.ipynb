````python
from transformers import AutoModel, AutoTokenizer

filepath_model = "C:/Users/biadge/MiniCPM-o-2_6"  # path to the model
model = AutoModel.from_pretrained(filepath_model, 
                                  trust_remote_code=True, 
                                  use_safetensors=True,
                                  attn_implementation='sdpa', 
                                  torch_dtype=torch.bfloat16, 
                                  init_vision=True,
                                  init_audio=True,
                                  init_tts=True)

model = model.eval().cuda()

tokenizer = AutoTokenizer.from_pretrained(filepath_model, trust_remote_code=True)
model.init_tts()
````

````python
from PIL import Image
import ast

def run_extract_concepts_local_llm(model, tokenizer, file) -> dict | str:

    with open("concept_extract_prompt.txt", "r") as file:
        prompt = file.read().strip()

    try:
        image = Image.open(file).convert('RGB')

        question = prompt
        msgs = [{'role': 'user', 'content': [image, question]}]
        res = model.chat(
            msgs=msgs,
            tokenizer=tokenizer
        )
    except Exception as e:
        return str(e)

    # remove the ```json from the response
    res = res.replace("```json", "")
    res = res.replace("```", "")

    # remove the first line
    res = res[res.find('\n') + 1:]

    # remove the last line
    res = res[:res.rfind('\n')]

    try:
        # convert string into dictionary
        res = dict(ast.literal_eval(res))
    except Exception as e:
        return str(res)

    return res
````

````python
# save json string into a json file

import json

with open("uml_example.json", "w") as file:
    json.dump(run_extract_concepts_local_llm(model, tokenizer, "uml_example.png"), file)
````

````python
def validate_concepts(model, tokenizer, file) -> dict:

    with open("validate_concept_prompt.txt", "r") as prompt_file:
        prompt = prompt_file.read().strip()

    # read file as string
    with open(file, "r") as file:
        file_str = file.read()

    try:
        question = prompt
        msgs = [{'role': 'user', 'content': [question, file_str]}]
        res = model.chat(
            msgs=msgs,
            tokenizer=tokenizer
        )
    except Exception as e:
        print(f"Error with model response: {e}")
        return res

    try:
        # convert string into dictionary
        res = dict(ast.literal_eval(res))
    except Exception as e:
        print(f"Error with converting dict: {e}")
        return res

    return res
````

````python
import json
import os
from json_repair import repair_json

def fix_malformed_json(json_string):
    json_string = json_string.strip()  # Remove any leading/trailing spaces
    if not json_string.startswith("{"):
        json_string = "{" + json_string  # Add missing opening brace
    if not json_string.endswith("}"):
        json_string = json_string + "}"  # Add missing closing brace

    return json_string

def fix_json(json_string):
    json_string = json_string.strip()  # Remove unwanted whitespace

    # Ensure JSON starts with { and ends with }
    if not json_string.startswith("{"):
        json_string = "{" + json_string
    if not json_string.endswith("}"):
        json_string = json_string.rstrip(",") + "]}]}"  # Close any missing brackets

    return json_string

for content in zip(files_to_validate,parsed_files_content):
    try:
        cleaned_content = content[1]
        # Remove newlines tabs and spaces
        cleaned_content = cleaned_content.replace("\\n", "").replace(" ", "")
        # Remove unnecessary escape characters
        cleaned_content = cleaned_content.replace("\\", "")
        # Remove first and last characters if they are not curly braces
        if not cleaned_content.startswith("{"):
            cleaned_content = cleaned_content[1:]
        if not cleaned_content.endswith("}"):
            cleaned_content = cleaned_content[:-1]
        # Remove unnecessary escape characters and ensure valid JSON format
        cleaned_content = fix_json(cleaned_content)
        # Repair any malformed JSON content
        cleaned_content = repair_json(cleaned_content)
        # Parse to ensure it's valid JSON
        parsed_content = json.loads(cleaned_content)
        # save the parsed content as a JSON file in the 'parsed_concepts' folder
        os.makedirs("validated_concepts_extracted", exist_ok=True)
        with open(os.path.join("validated_concepts_extracted", f"{os.path.basename(content[0])}.json"), "w") as file:
            json.dump(parsed_content, file)
            #file.write(cleaned_content)

        print(f"Successfully cleaned and saved content for {content[0]}")
    except Exception as e:
        print(f"Error cleaning content: {e}")
        print(f"File: {content[0]}")
        continue
````
