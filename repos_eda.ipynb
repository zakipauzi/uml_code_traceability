```python
import pandas as pd

repos_df = pd.read_csv("repo_stargazers_count.csv")
repos_df
```

```python
repos_df = repos_df[:100]
repos_df[:10]
```

```python
uml_files = pd.read_csv("UMLFiles_List_V2.0.csv")
uml_files["Project"] = uml_files["Project"].apply(lambda x: x[:-1])
uml_files
```

```python
proj_df = pd.read_csv("Project_FileTypes_V2.0.csv")
proj_df["Repo_Name"] = proj_df["Repo_Name"].apply(lambda x: x[:-1])
proj_df
```

```python
df = repos_df.merge(proj_df, left_on='repo', right_on='Repo_Name').merge(uml_files, left_on='repo', right_on='Project')
```

```python
df = df.drop(columns=['Repo_Name','Project'])
df.head()
```

```python
df.to_csv("repos_df.csv", index=False)
```
