```python
import requests
from dotenv import load_dotenv
import os

_ = load_dotenv()

GH_USER = os.environ["GH_USERNAME"]
GH_TOKEN = os.environ["GH_TOKEN"]
```

```python
def get_star_count(repo_name):
    r = requests.get('https://api.github.com/repos/{}'.format(repo_name), auth=(GH_USER,GH_TOKEN), timeout=10)
    
    print(r.status_code)
    if r.status_code==200:
        return r.json()['stargazers_count']
    else:
        return None
```

## concatenate the stargazer counts

```python
# will be merging 1 to 17 parts

prefix_filename = "repo_stargazers_count_"
files = []

for i in range(1,18):
    curr_file = prefix_filename + str(i)
    print("downloading csv file {}".format(curr_file))
    
    files.append(pd.read_csv(curr_file + ".csv"))
```

```python
stars_df = pd.concat(files, ignore_index=True)
stars_df
```

```python
#remove na, sort values

stars_df = stars_df.dropna()
stars_df = stars_df.sort_values(by='stargazers_count', ascending=False).reset_index(drop=True)
stars_df
```

```python
stars_df.to_csv('repo_stargazers_count.csv', index=False, encoding='utf-8')
```
